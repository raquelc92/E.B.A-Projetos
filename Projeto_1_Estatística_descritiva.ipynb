**Parte 3: Correlações**


Qual é a relação de estado civil com número de filhos? Será que as pessoas casadas têm um maior número de filhos? Filhos está na coluna kids 

As pessoas gastam mais ou menos em nossa plataforma quando têm filhos? Veja nas colunas expenses e kids 

Pessoas que têm um maior salário gastam mais? Veja nas colunas Income e expenses 


```python
df.groupby("marital_status")['kids'].mean()
```

```python
fig2 = sns.boxplot(x="marital_status", y="kids", data=df)
plt.xticks(rotation=90);
```

```python
M = sns.FacetGrid(df, col="marital_status")
M.map(sns.histplot, "kids")
plt.show()
```

Verificamos que a maioria dos clientes possui 1 filho, independente do estado civil

```python
#Correlação entre gastos e quantidade de filhos
df.groupby("kids")['expenses'].mean()
```

```python
sns.scatterplot(x="expenses", y="kids", data=df)
```

Os dados demonstram que os clientes que não possuem filhos tendem a gastar mais

```python
#Correlação entre nível de salários e consumo na plataforma
sns.scatterplot(x="expenses", y="Income", data=df)
```

```python
plt.figure(figsize=(20,12)) 
sns.lmplot(x ='expenses', y ='Income', data = df,scatter_kws={'alpha':0.5,'color':'red'})
plt.xlabel("Expenses (dollars)")
plt.ylabel("Income (dollars)")
plt.xticks(rotation=45)
plt.title("Correlation between expenses x income")
plt.show()
```

Os dois gráficos acima demonstram uma correlação positiva entre salário e consumo dentro da plataforma. Existe uma tendência dos clientes que ganham salários maiores a gastarem mais no ifood.
